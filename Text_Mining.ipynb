```python
import textract
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

nltk.download('all')
```

# Convertendo de PDF para txt

```python
text = textract.process("datasets/Texto_03_Exclusao_Digital.pdf", encoding="utf-8")
text = text.decode("utf-8")

print(text)
```

# Processando o texto

```python
import re

def process_text(text):

    # Tokenização
    tokens = word_tokenize(text.lower())

    # Remove stop words em português e remove pontuações
    filtered_tokens = [token for token in tokens if token not in stopwords.words('portuguese') and not re.search(r'[.,;:!?()\[\]{}"\'’“”]', token)]

    # Lematização (busca reduzir as palavras em sua raiz)
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Converte os tokens para string
    processed_text = ' '.join(lemmatized_tokens)

    return processed_text

text = process_text(text)
print(text)
```

```python
# Criar um objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Exibir a nuvem de palavras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
